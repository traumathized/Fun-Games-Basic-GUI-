```python
!pip install ipywidgets

import random
import ipywidgets as widgets
from IPython.display import display


# Scores
user_score = 0
computer_score = 0

# Output display
output = widgets.Output()

# Labels
score_label = widgets.Label(value="Score â†’ You: 0 | Computer: 0")
result_label = widgets.Label(value="Click Roll Dice to start!")

# Button
roll_button = widgets.Button(description="ðŸŽ² Roll Dice")

def roll_dice(b):
    global user_score, computer_score

    user_roll = random.randint(1, 6)
    computer_roll = random.randint(1, 6)

    if user_roll > computer_roll:
        user_score += 1
        result = "ðŸŽ‰ You Win This Round!"
    elif computer_roll > user_roll:
        computer_score += 1
        result = "ðŸ’» Computer Wins This Round!"
    else:
        result = "ðŸ¤ It's a Tie!"

    score_label.value = f"Score â†’ You: {user_score} | Computer: {computer_score}"
    result_label.value = f"You rolled {user_roll} | Computer rolled {computer_roll} â†’ {result}"

roll_button.on_click(roll_dice)

display(score_label)
display(result_label)
display(roll_button)
```
